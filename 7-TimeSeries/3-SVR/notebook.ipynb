```python
import os
import warnings
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime as dt
import math

from sklearn.svm import SVR
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("Month_Value_1.csv")
df = df.set_index("Period")
df.head(10)
```

```python
train_start_dt = "01.01.2015"
test_start_dt = "01.10.2015"

train = df.copy()[(df.index >= train_start_dt) & (df.index < test_start_dt)][['Sales_quantity']]
test = df.copy()[df.index >= test_start_dt][['Sales_quantity']]

print('Training data shape: ', train.shape)
print('Test data shape: ', test.shape)
```
